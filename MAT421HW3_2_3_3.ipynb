## 3.2 - Continuity and Differentiation

One of the most important concepts to learn in Calculus and Mathmatical Analysis would be Limits. Limits help define Conitunity, derivatives, and intergrals. A function that has **Conitunity** is a function that doesn't contain any sudden changes to the function. **Differentiations** are operations that are used to describe a rate of change in a function. Operations such as Derivatives and Intergrals are considered **Differentiations**.

```python
## Continuos Function Example

import numpy as np
import matplotlib.pyplot as plt

def my_continuous_function(x):
    return np.cos(x)

x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_values = my_continuous_function(x_values)


plt.plot(x_values, y_values)
plt.title('Continuous Function Example')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
