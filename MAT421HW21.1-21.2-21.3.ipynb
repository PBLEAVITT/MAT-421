## 21.2 - Riemanns Integral

**Riemanns Integral** is the simplest method for approximating integrals. Its done by summing up the area of rectangles that are defined by subintervals.

**Example:** Use the left Riemann Integral, right Riemann Integral, and Midpoint Rule to approximate the following function with 16 evenly spaced grid points over the whole interval. Compare this value to the exact value of 4.

```python
import numpy as np

a = 0
b = np.pi
n = 16
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 4 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 4 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] \
        + x[1:])/2))
err_mid = 4 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

```python
import numpy as np

a = 0
b = np.pi
n = 16
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_trap = (h/2)*(f[0] + \
          2 * sum(f[1:n-1]) + f[n-1])
err_trap = 4 - I_trap

print(I_trap)
print(err_trap)
```
