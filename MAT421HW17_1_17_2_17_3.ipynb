# *17.1 - Interpolation Problem Statement*

**Interpolation** is the use a set of independent data values and a set of dependent data values to estimate an unknown value within a data set. This is done with the use of a **Interpolation Function**. Interpolation doesn't require an underlying model for the data, but it is important to state the quality of the interpolation as either 'smooth' or 'jagged'. There are several common Interpolation methods such as Liner, Cubic-spline, Lagrange Polynomial, and Newton's Polynomial.

**Visual representation:**

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 1, 2, 6, 8]
y = [1, 3, 2, 5, 4]

plt.figure(figsize = (10,8))
plt.plot(x, y, 'ob')
plt.plot(4,3.5, 'ro')
plt.title('Interpolation Problem Statement')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# *17.2 - Linear Interpolation*

**Linear Interpolation** assumes that the estimated point is on line that joins the nearest points to the left and right.

As an Engineering student I have had to untilze linear interpolation alot in classes like theromodynamics, fluids, and heat transfer. When I perform linear interpolation I think of it in terms of percents. For example, if I have a data set that has values of 50 and 100 and another that has 7 and 8 and I need to find a value from the first data set using 7.7 from the second data set. The first question I ask myself is what percentage is 7.7 relative to 7 and 8? The answer would be 70%. Then I take the difference from the first set (100-50=50) then ask what is 70% of 50? The answer would be 35. Then finally I add that value to the value of the lowest point on the first data set (50) so I would add 50+35 and get 85 and that would be my answer.

**Visualized Example:** When Saturated Water has a pressure of 75 kPa the saturated temperature is 91.76 degrees celcius and when it has a pressure of 100 kPa the saturated temperature is 99.61 dgrees celcius. What is the temperature of Saturated Water when it has a pressure of 87 kPa?

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [75, 100]
y = [91.76, 99.61]

f = interp1d(x, y)
x_hat = f(87)
print(x_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(87, x_hat, 'ro')
plt.title('Linear Interpolation at y = 87')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# *17.3 - Cubic Spline Interpolation*

**Cubic Spline Interpolation** uses a set of piecewise cubic functions. This causes the line from one point to another to be parabalic in shape.

**Visualized Example:** Consider the data set x= [0,3,5,7] and y=[1,4,2,8] for 0<=x<=7.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 3, 5, 7]
y = [1, 4, 2, 8]

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 7, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
